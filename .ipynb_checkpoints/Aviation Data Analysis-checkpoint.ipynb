# Aviation Data Analysis

With hundreds of thousands of flights every day, the aviation industry offers huge opportunities but carries significant risks. Understanding which aircraft types are safer can help companies make smarter, data-driven decisions. In this analysis, we explore patterns in flight incidents to uncover aircraft associated with lower risk and higher reliability.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('AviationData.csv', encoding='latin1')
data.head()
```

```python
data.shape
```

## Data Preparation and Cleaning

Making the column names consistent

```python
data.columns = data.columns.str.lower().str.replace('.', '_')
data.columns
```

```python
data.isna().sum()
```

Every column has a missing value except 4 four columns.

We will drop any column that has more than 20000 missing values. So that it can be easier to deal the missing values.

```python
selected_cols = []
for col in data.columns:
    if data[col].isna().sum() < 20000:
        selected_cols.append(col)

selected_cols                             
```

```python
data = data[selected_cols]
data.head()
```

Drop any missing rows in the `total_fatal_injuries` column. Because if this column has a missing value it's also that the other column has missing values.

```python
print(data.shape)
data = data.dropna(subset=['total_fatal_injuries'])
print(data.shape)
```

```python
data.head()
```

- Filling in these columns `total_serious_injuries`, `total_minor_injuries`, `total_uninjured` with 0's because originally they were left blank implying that nothing happened
- Filling `make` and `model` with **`Unknown`** because maybe the reports was unable identify it

```python
print(len(data[data['total_serious_injuries'] == 0]))

# Filling these columns with 0, because they were left blank implying that there was nothing that happened
cols = ['total_serious_injuries', 'total_minor_injuries', 'total_uninjured']
for col in cols:
    data[col] = data[col].fillna(0)
```

```python
data['make'] = data['make'].fillna('Unknown')
data['model'] = data['model'].fillna('Unknown')
data['amateur_built'] = data['amateur_built'].fillna(data['amateur_built'].mode()[0])
data['registration_number'] = data['registration_number'].fillna('None')
```

```python
data.isna().sum()
```

```python
data['injury_severity'].value_counts()
```

```python
pattern = r'\(\d+\)'
data['injury_severity'] = data['injury_severity'].replace(pattern, '', regex=True).str.strip()

data.loc[(data['injury_severity'].isna() & data['total_fatal_injuries'] > 0), 'injury_severity'] = 'Fatal'
data.loc[data['injury_severity'].isna() & (data['total_fatal_injuries'] == 0),'injury_severity'] = 'Non-Fatal'
```

- The column `injury_severity` has many duplicates.
- Cleaning the missing values in this `injury_severity` according to the `total_fatal_injuries` column.

```python
cols = ['location', 'country']
for col in cols:
    data[col] = data[col].fillna('Unknown')
```

```python
data['number_of_engines'] = data['number_of_engines'].fillna(data.groupby(['make', 'model'])['number_of_engines'].transform('median'))
data['number_of_engines'] = data['number_of_engines'].fillna(data['number_of_engines'].median())
```

```python
data['aircraft_damage'].value_counts()
```

```python
data = data.drop('publication_date', axis=1)
cols = ['aircraft_damage', 'engine_type', 'purpose_of_flight', 'weather_condition', 'report_status']
for col in cols:
    data[col] = data[col].fillna('Unknown')
```

- Filling the rest of the columns with `Unknown`. Because in Aviation most rows are left undocumented meaning the cause might be truly undeteremined.
- Dropping the `publication_date` columns since we already have `event_date` to use.

```python
data.isna().sum()
```

```python
data['event_date'] = pd.to_datetime(data['event_date'])
```

```python
data['make'] = data['make'].str.lower().str.strip().str.title()
```

```python
print(data.shape)
balloon_planes = data[data['report_status'].str.contains('balloons?', regex=True, na=False, case=False)]
data = data.drop(balloon_planes.index)
helis = data[data['report_status'].str.contains('helicopter', regex=True, na=False, case=False)]
data = data.drop(helis.index)
print(data.shape)
```

```python
tobe_dropped = ['ballons?', 'helicopter']
for plane in tobe_dropped:
    plane_type = data[data['report_status'].str.contains(plane, regex=True, na=False, case=False)]
    data = data.drop(plane_type.index)
```

Dropped `ballons` and `helicopters` since they are not considered as airplanes.

```python
data['injury_severity'] = data['injury_severity'].str.replace('Minor', 'Non-Fatal').str.replace('Serious', 'Fatal').str.replace('Incident', 'Non-Fatal')
data = data[data['injury_severity'] != 'Unavailable']
data['injury_severity'].value_counts()
```

```python
data = data[data['number_of_engines'] >= 1]
```

## Data Analysis

## Top 50 Best Aircrafts

We will have to narrow down this list so that we get the manufacturers that produce the most.

```python
top_50 = list(data['make'].value_counts()[:50].index)
data_top_50 = data[data['make'].isin(top_50)]
data_top_50.shape
```

```python
crash_severity = data_top_50.groupby(['make', 'injury_severity']).size().unstack()
crash_severity['total_crashes'] = crash_severity['Fatal'] + crash_severity['Non-Fatal']
crash_severity['fatal_crash_ratio'] = crash_severity['Fatal'] / crash_severity['Non-Fatal']
crash_severity = crash_severity[crash_severity['total_crashes'] > 800]
crash_severity.sort_values('fatal_crash_ratio')
```

```python
lowest_crash = crash_severity.sort_values('fatal_crash_ratio', ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=lowest_crash.index, y=lowest_crash['fatal_crash_ratio'], palette='magma', hue=lowest_crash.index, legend=False)
plt.xlabel('Aircraft Make')
plt.ylabel('Fatal Crash Ratio')
plt.title('Top aircraft makes with the lowest aircraft crash rates')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

- The data further narrowed down to manufacturers who have more than 800 products.
    - `Boeing`, `Grumman` and `Cessna` were the top three airplanes that had the least crashes.

```python
data['decade'] = (data['event_date'].dt.year // 10) * 10
data.groupby('decade').size().plot(kind='line', marker='o')
plt.title('Crashes Over Decades')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
data.groupby('decade')['total_fatal_injuries'].sum().plot(kind='line', marker='o', color='red')
plt.title('Fatalities Over Decades')
plt.ylabel('Total Fatal Injuries')
plt.show()
```

- Fatalities and number of injuries decreased over the years meaning the modern aviation has become exponential safer in the 21st century.

## Number of Engines with the least crashes

```python
data_top_50.head()
```

```python
data_top_50['number_of_engines'] = pd.to_numeric(data_top_50['number_of_engines'], errors='coerce')

engine_data = data_top_50.dropna(subset=['number_of_engines'])
engine_crashes = engine_data.groupby('number_of_engines').size().sort_values()

least_crashes = engine_crashes[engine_crashes == engine_crashes]
print("Number of engines with the least crashes:")
print(least_crashes)


engine_crashes.plot(kind='bar', figsize=(8,5), title='Crashes by Number of Engines')
plt.xlabel('Number of Engines')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
data_top_50['decades'] = (data_top_50['event_date'].dt.year // 10) * 10
```

```python
decade_counts = data_top_50.groupby(['decades','number_of_engines']).size().reset_index(name='crashes')

for engines in sorted(decade_counts['number_of_engines'].unique()):
    subset = decade_counts[decade_counts['number_of_engines']==engines]
    plt.scatter(subset['decades'], subset['crashes'], label=f'{engines} engines')

plt.xlabel('Decade')
plt.ylabel('Number of Crashes')
plt.title('Crashes by Number of Engines Over Decades')
plt.legend()
```

- Single-Engine - `50,602` crashes: This massive number reflects the vulnerability of general aviation. According to the [FAA](https://armoneyandpolitics.com/arkansas-mistakes-plane-crashes/), nearly four aircraft crash daily, and almost all are single-engine planes. if the only engine fails, the pilot must glide to a landing, leading to a much higher accident rate.
- Four-Engine - `391` crashes: This type of aircraft has the least aircrashes because they are rare and small in number. Mostly used for heavy cargo.
- Three-Engine - `434` crashes: This Tri-engine aircraft types have become rare in use thefore decreasing its crashes.
- Two-Engine - `9544` crashes: The Twin-engine aircraft is considered safer than single and more common than both three and four engine type aircrafts.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data_top_50, x='aircraft_damage', y='total_fatal_injuries', hue='number_of_engines', palette='RdBu')
plt.xticks(rotation=45)
plt.title('Fatal Injuries by Aircraft Damage')
plt.show()
```

- Two or more number of engine are rare to crash but once they crash they can have very fatal number of injuries

**Recommendation**:
1. Focus on safer Aircraft makes:
    - Prioritize aircraft with lower accident rates for purchase. This reduces the risk exposure in the industry while staying in the market for long periods
2. Invest in 2 or more engine aircrafts:
    - Emphasize on multi-engine aircrafts with high maintenance. This reduces the insurance cost and improve customer confidentiality.
